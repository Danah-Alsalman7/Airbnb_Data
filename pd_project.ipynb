##Data Loading and Initial Exploration

First we need to load the selected dataset using pandas
we have Tow datasets we will use merge() to mergeing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('/content/Airbnb_Data/Listings.csv',encoding = 'unicode_escape')
df2 = pd.read_csv('/content/Airbnb_Data/Reviews.csv')
df = pd.merge(df1, df2, on='listing_id', how='inner')
df
```

```python
#number of rows and columns,

rows = len(df.axes[0])
columns = len(df.axes[1])
print("Number of Rows: ", rows)
print("Number of Columns: ", columns)

df.dtypes

df.info()
df.shape
```

```python
# The first few rows of the dataset

df.head()
```

##Data Cleaning


```python
#we take a seed sample of tha data
np.random.seed(1)
dfs = df.sample(70000)
```

```python
# # Missing values

# Impute missing 'name' values with "Unknown"
dfs['name'].fillna('Unknown', inplace=True)

# # Impute missing 'host_since' values with "Unknown"
dfs['host_since'].fillna('Unknown', inplace=True)

# # Dealing with 'host_response_time' missing valuse
mode_response_time = dfs['host_response_time'].mode()[0]
dfs['host_response_time'].fillna(mode_response_time, inplace=True)

# # Impute missing values in 'host_is_superhost' with False
dfs['host_is_superhost'].fillna(False, inplace=True)

# # Dealing with 'host_total_listings_count' missing valuse
mean_listings_count = dfs['host_total_listings_count'].mean()
dfs['host_total_listings_count'].fillna(mean_listings_count, inplace=True)

# # Impute missing values in 'host_has_profile_pic' with False
dfs['host_has_profile_pic'].fillna(False, inplace=True)

# # Impute missing values in 'host_identity_verified' with False
dfs['host_identity_verified'].fillna(False, inplace=True)

# # Impute missing 'district' values based on 'neighbourhood'
dfs['district'] = dfs['district'].fillna(dfs['neighbourhood'])

# # Impute missing values in 'bedrooms' with zero
dfs['bedrooms'].fillna(0, inplace=True)

# # Dealing with 'review_scores_rating' missing valuse
mean_scores_rating = dfs['review_scores_rating'].mean()
dfs['review_scores_rating'].fillna(mean_scores_rating, inplace=True)

# # Dealing with 'review_scores_accuracy' missing valuse
median_review_scores_accuracy = dfs['review_scores_accuracy'].median()
dfs['review_scores_accuracy'].fillna(median_review_scores_accuracy, inplace=True)

# # Dealing with 'review_scores_cleanliness' missing valuse
median_review_scores_cleanliness = dfs['review_scores_cleanliness'].median()
dfs['review_scores_cleanliness'].fillna(median_review_scores_cleanliness, inplace=True)

# # Dealing with 'review_scores_checkin' missing valuse
median_review_scores_checkin = dfs['review_scores_checkin'].median()
dfs['review_scores_checkin'].fillna(median_review_scores_checkin, inplace=True)

# # Dealing with 'review_scores_communication' missing valuse
median_review_scores_communication = dfs['review_scores_communication'].median()
dfs['review_scores_communication'].fillna(median_review_scores_communication, inplace=True)

# # Dealing with 'review_scores_location' missing valuse
mean_review_scores_location = dfs['review_scores_location'].mean()
dfs['review_scores_location'].fillna(mean_review_scores_location, inplace=True)

# # Dealing with 'review_scores_value' missing valuse
mean_review_scores_value = dfs['review_scores_value'].mean()
dfs['review_scores_value'].fillna(mean_review_scores_value, inplace=True)

dfs.isnull().sum()
```

```python
# Duplicate values
dfs.duplicated().sum()
```

```python
# Converting data type
# dfs.dtypes

# Converting 'host_since' column to datetime type
df['host_since'] = pd.to_datetime(df['host_since'],errors='coerce')

# Converting 'date' column to datetime type
dfs['date'] = pd.to_datetime(dfs['date'])

# Converting 'host_identity_verified' column to datetime type
dfs['host_identity_verified'] = dfs['host_identity_verified'].replace({'t': True, 'f': False})
dfs['host_identity_verified'] = dfs['host_identity_verified'].astype(bool)

# Converting 'host_has_profile_pic' column to datetime type
dfs['host_has_profile_pic'] = dfs['host_has_profile_pic'].replace({'t': True, 'f': False})
dfs['host_has_profile_pic'] = dfs['host_has_profile_pic'].astype(bool)

# Converting 'instant_bookable' column to datetime type
dfs['instant_bookable'] = dfs['instant_bookable'].replace({'t': True, 'f': False})
dfs['instant_bookable'] = dfs['instant_bookable'].astype(bool)

dfs.dtypes
```

##Feature Engineering
